```python
import pandas as pd
import numpy as np
import nltk
import re
import tokenizers
import emoji,string
```

```python
def remove(text):
    text = re.sub(r"http\S+", "", text)#removing the URLs
    text = re.sub("@.*", "USER", text)#substituting mentions with user
    text  = "".join([char for char in text if char is not emoji.is_emoji(char)])#emoji removal
    text  = "".join([char for char in text if char not in string.punctuation])#removing punctuation
    text = re.sub('[0-9]+', '', text)
    return text
```

```python
df = pd.read_csv("psuedo_data.csv")
df.head()
```

```python
df['clean_comment']=df['text'].apply(lambda x:remove(x.lower()))
df['Tokens']=df['clean_comment'].apply(lambda x:nltk.tokenize.wordpunct_tokenize(x))
# df['clean_comment']=df['Tokens'].apply(lambda x: stopwords_removal(x))
# df['Tokens']=df['clean_comment'].apply(lambda x:nltk.tokenize.wordpunct_tokenize(x))
df['Length']=df['Tokens'].apply(lambda x:len(x))
df.to_csv('Cleaned_df.csv',index=False)
```

```python
def crude_tltrt(text):
  return transliterate(text, sanscript.ITRANS, sanscript.DEVANAGARI)

from elt import translit

def tltrt(text):
  to_hindi = translit('hindi')
  return to_hindi.convert([text])[0]

def lambda_tltrt(text):
  try:
    return tltrt(text)
  except:
    return crude_tltrt(text)
```

```python
nltk.download('words')
def combined_transliterate(text):
  print(text)
  res = ""
  english_words = set(words.words())
  for word in text.split():
    if word in english_words:
      res += " "
      res += word
    else:
      res += " "
      res += lambda_tltrt(word)

  return res.strip()

combined_transliterate("knowing ki vikas kitna samjhata hai priyanka aur itch guard luv ko usne bola tha ben wali baat ab sallu ne bhi agree kiya")
```

```python
df["combined_text"] = df["clean_comment"].apply(lambda x: combined_transliterate(x))
```
